```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

## TextBlob

```python
from textblob import TextBlob
```

```python
print(sentence_1)
sentiment_score_1 = TextBlob(sentence_1)
print(sentiment_score_1.sentiment) # Sentiment(polarity=0.525, subjectivity=0.875)
```

```python
print(sentence_2)
sentiment_score_2 = TextBlob(sentence_2)
print(sentiment_score_2.sentiment)
```

```python
print(sentence_3)
sentiment_score_3 = TextBlob(sentence_3)
print(sentiment_score_3.sentiment)
```

```python
print(sentence_4)
sentiment_score_4 = TextBlob(sentence_4)
print(sentiment_score_4.sentiment)
```

## VADER

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
vader_sentiment = SentimentIntensityAnalyzer()
```

```python
print(sentence_1)
print(vader_sentiment.polarity_scores(text=sentence_1))
```

```python
print(sentence_2)
print(vader_sentiment.polarity_scores(text=sentence_2))
```

```python
print(sentence_3)
print(vader_sentiment.polarity_scores(text=sentence_3))
```

```python
print(sentence_4)
print(vader_sentiment.polarity_scores(text=sentence_4))
```

## Transformers

```python
import transformers
from transformers import pipeline

sentiment_pipeline = pipeline("sentiment-analysis")
# Pipeline allows you to use powerful pre-trained models without needing to build from scratch.
```

```python
print(sentence_1)
sentiment_pipeline(sentence_1) # [{'label': 'POSITIVE', 'score': 0.9998176693916321}]
```

```python
print(sentence_2)
sentiment_pipeline(sentence_2)
```

```python
print(sentence_3)
sentiment_pipeline(sentence_3)
```

```python
print(sentence_4)
sentiment_pipeline(sentence_4) # [{'label': 'POSITIVE', 'score': 0.9802700281143188}]
# Mis-classified. 
# Options: (1) use different pre-trained models.
```

```python
specific_model = pipeline('sentiment-analysis', model="finiteautomata/bertweet-base-sentiment-analysis")
```

```python
print(sentence_4)
specific_model(sentence_4)
```

```python
print(sentence_1)
specific_model(sentence_2)
```

```python
print(sentence_2)
specific_model(sentence_2)
```

```python
print(sentence_3)
specific_model(sentence_3)
```
