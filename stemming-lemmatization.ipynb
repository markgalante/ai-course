```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem("running")
```

```python
connect_words = ["connection", "connected", "connecting", "connections", "connectivity"]
for word in connect_words:
    print(f"{word} --> {ps.stem(word)}")
```

```python
learn_words = ["learn", "learnability", "learned", "learning", "learns", "learner", "learns"]
for word in learn_words:
    print(f"{word} --> {ps.stem(word)}")
```

```python
improper_words = ["studies", "worse", "likes", "better"]
for word in improper_words:
    print(f"{word} --> {ps.stem(word)}")
```
