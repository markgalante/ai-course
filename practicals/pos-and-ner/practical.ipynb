```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bbc_data = pd.read_csv("bbc_news.csv")
```

```python
bbc_data.head()
```

```python
bbc_data.info()
```

```python
titles = pd.DataFrame(bbc_data["title"])
titles.head()
```

```python
# Clean up titles
titles["lowercase"] = titles["title"].str.lower()
titles.head()
```

```python
en_stopwords = stopwords.words("english")
titles["no_stopwords"] = titles["lowercase"].apply(lambda words: ' '.join([word for word in words.split() if word not in (en_stopwords)]))
```

```python
titles["no_stopwords_no_punc"] = titles.apply(lambda row: re.sub(r"[^\w\s]", "", row["no_stopwords"]), axis=1)
# `axis = 1` means that this is applies across rows and not columns.
```

```python
# Tokenise clean and raw titles
titles['tokens_raw'] = titles.apply(lambda row: word_tokenize(row['title']), axis = 1)
titles['tokens_clean'] = titles.apply(lambda row: word_tokenize(row['no_stopwords_no_punc']), axis = 1)
```

```python
# Lemmatizing
lemmatizer = WordNetLemmatizer()
titles['tokens_clean_lemmatized'] = titles['tokens_clean'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
titles.head()
```

```python
tokens_raw_list = sum(titles["tokens_raw"], [])
tokens_clean_list = sum(titles["tokens_clean_lemmatized"], [])
```

### POST tagging

```python
# Initialize SpaCy model that'll be used
nlp = spacy.load('en_core_web_sm')
```

```python
# Contains text and text annotations
spacy_doc = nlp(" ".join(tokens_raw_list))
```

```python
pos_df = pd.DataFrame(columns=["token", "pos_tag"])

for token in spacy_doc:
    pos_df = pd.concat([pos_df, pd.DataFrame.from_records([{"token": token.text, "pos_tag": token.pos_}])], ignore_index=True)
```

```python
pos_df_counts = pos_df.groupby(["token", "pos_tag"]).size().reset_index(name="counts").sort_values(by="counts", ascending=False)
# grouped by 2 columns (token and tag)
# `size()` counts how many rows fall into each of the groups. IOW, how often each token-tag pair occurs.
# `reset_index()` removes group labels from the index and turns them into regular columns with a column name ("counts")
pos_df_counts.head(10)
```

```python
nouns = pos_df_counts[pos_df_counts.pos_tag == "NOUN"][0:10]
nouns
```

```python
verbs = pos_df_counts[pos_df_counts.pos_tag == "VERB"][0:10]
verbs
```

```python
adjective = pos_df_counts[pos_df_counts.pos_tag == "ADJ"][0:10]
adjective
```

### NER
